```python
class Cohort:
  def __init__(self, name):
    self.name = name

dict_dir = r'C:\Users\neil_\DellXPS_June2024\OneDrive\Desktop\New UHN\Precision Medicine\carotid_plaque\data'
```

```python
Asymp = Cohort("Asymptomatic")

with open(dict_dir+'/as_mirna_targets.pkl', 'rb') as file:
    Asymp.mirna_target_map = pickle.load(file)

# List of all targeted genes
Asymp.mirna_targets = list(set([gene for gene_list in list(Asymp.mirna_target_map.values()) for gene in gene_list]))

# Get background gene list, use all miRNA targets from MEINTURNET
file_path = r'C:\Users\neil_\DellXPS_June2024\OneDrive\Desktop\New UHN\Precision Medicine\carotid_plaque\data\Mienturnet Enrichment results miRTarBase_Asymptomatic_PlaquevsMarginal_June2024.xlsx'
df = pd.read_excel(file_path, header=0, skiprows=[0])
Asymp.bg_set = list(df["Gene Symbol"])

Asymp.gseapy = {}
```

```python
Symp = Cohort("Symptomatic")

with open(dict_dir+'/s_mirna_targets.pkl', 'rb') as file:
    Symp.mirna_target_map = pickle.load(file)

# List of all targeted genes
Symp.mirna_targets = list(set([gene for gene_list in list(Symp.mirna_target_map.values()) for gene in gene_list]))

# Get background gene list, use all miRNA targets from MEINTURNET
file_path = r'C:\Users\neil_\DellXPS_June2024\OneDrive\Desktop\New UHN\Precision Medicine\carotid_plaque\data\Mienturnet Enrichment results miRTarBase_Symptomatic_PlaquevsMarginal.xlsx'
df = pd.read_excel(file_path, header=0, skiprows=[0])
Symp.bg_set = list(df["Gene Symbol"])

Symp.gseapy = {}
```

```python
def add_fold_enrichment(target_enr_df, bg_enr_df, n_DEGs, n_genes):
    ''' Fold enrichment is calculated as FE = (a/b) / (c/d) where
    a = # genes in pathway from gene list
    b = # genes in gene list of interest
    c = # genes in pathway from bg set
    d = # genes in bg set'''
    bg_enr_df['c'] = bg_enr_df["Overlap"].str.split("/").str[0].astype(int)
    new_df = pd.merge(target_enr_df, bg_enr_df[["Term", "c"]], how='left', on="Term")

    new_df['a'] = new_df["Overlap"].str.split("/").str[0].astype(int)
    new_df['b'] = n_DEGs
    new_df['d'] = n_genes
    new_df["Fold_Enrichment"] = (new_df['a']/ new_df['b']) / (new_df['c'] /new_df['d'])

    return new_df
```

```python
for Cohort in [Asymp, Symp]:
    full_target_list = Cohort.mirna_targets
    bg_set = Cohort.bg_set
    for db in ['KEGG_2021_Human', 'GO_Biological_Process_2023']:
        g = gseapy.enrichr(gene_list=full_target_list,
                       gene_sets=db,
                       organism='human',
                      background =bg_set)
        enriched_pthwys_df = g.results
        filtered_pthwys_df = enriched_pthwys_df[enriched_pthwys_df["Adjusted P-value"]<0.05]
        
        #Cohort.gseapy[db] = filtered_pthwys_df

        # Fold Enrichment calculation
        g2 = gseapy.enrichr(gene_list=bg_set,
                       gene_sets=db,
                       organism='human')
        full_df = g2.results # Need this to get # of bg genes in pathways

        n_DEGs = len(Cohort.mirna_targets)
        n_bg_genes = len(Cohort.bg_set)
        df = add_fold_enrichment(filtered_pthwys_df, full_df, n_DEGs, n_bg_genes)
        df["cohort"] = Cohort.name
        Cohort.gseapy[db] = df
```

```python
db = "GO_Biological_Process_2023"

# Compile pathways across both cohorts for db
concatenated_df = pd.concat([Asymp.gseapy[db], Symp.gseapy[db]])
concatenated_df['Term'] = concatenated_df['Term'].str.replace(r'\(GO:\d+\)', '', regex=True).str.strip()
pivot_df = concatenated_df.pivot_table(index='Term', columns='cohort', values='Fold_Enrichment')
#pivot_df = concatenated_df.pivot_table(index='Term', columns='cohort', values='Adjusted P-value')
pivot_df = pivot_df.fillna(1)
#pivot_df.to_csv(f"mirna_pathways_{db}.csv")
# Emphasize the pathways where FE differs the most
variation = pivot_df.std(axis=1)
# Sort bp_labels by variation in descending order and select top 30
top_30_bp_labels = variation.nlargest(30).index
# Filter pivot_df to keep only these top 30 bp_labels
filtered_df = pivot_df.loc[top_30_bp_labels]

# Plot 30 pathways heatmap
f = sns.clustermap(filtered_df,cmap=cm.batlow, figsize=(8, 10), annot=False, row_cluster=False, col_cluster=False)
#f = sns.clustermap(pivot_df, cmap=cm.batlow, figsize=(12, 12), annot=False)

plt.title(f'Overrepresentation Analysis of DEms against {db}')
plt.ylabel('Fold Enrichment')

#f.savefig(f"dem_{db}_overrep_heatmap.png")
plt.show()
```

```python
import pandas as pd
def process_tab(file_path, cohort_label, q=0.05):
    # Open the file and read the first line
    with open(file_path, 'r') as file:
        first_line = file.readline().strip()
    
    # Extract the values from the first line
    parts = first_line.split(',')
    DEPset_size = int(parts[0].split('=')[1])
    background_set_size = int(parts[1].split('=')[1])
    
    # Read table as pandas DataFrame
    raw_table = pd.read_csv(file_path, sep='\t', header=1)
    
    # Add a new column 'overlap_count' which is the count of semicolons in 'members_input_overlap' plus 1
    raw_table['overlap_count'] = raw_table['members_input_overlap'].str.count(';') + 1

    # Calculate 'fold_enrichment' and add as a new column
    raw_table['fold_enrichment'] = (raw_table['overlap_count'] / raw_table['size']) / (DEPset_size / background_set_size)
    
    # Add the 'cohort' column
    raw_table['cohort'] = cohort_label

    # Strip " - Homo sapiens (human)" from the "pathway" column
    raw_table['pathway'] = raw_table['pathway'].str.replace(" - Homo sapiens \(human\)", "", regex=True)

    # Update the boolean mask to include 'fold_enrichment' > 1
    mask = (raw_table['q-value'] < q) & (raw_table['fold_enrichment'] > 1)

    # Apply the mask to filter the table
    filtered_table = raw_table[mask]
    
    return filtered_table
```

```python
symp = process_tab(r'C:\Users\neil_\DellXPS_June2024\OneDrive\Desktop\New UHN\Precision Medicine\carotid_plaque\data\symp_dem_kegg_cdb.tab', 'symptomatic')
asymp = process_tab(r'C:\Users\neil_\DellXPS_June2024\OneDrive\Desktop\New UHN\Precision Medicine\carotid_plaque\data\asymp_dem_kegg_cdb.tab', 'asymptomatic')

concatenated_df = pd.concat([symp, asymp])
```

```python
df1 = concatenated_df.sort_values("fold_enrichment", ascending=False)
df1 = df1[["pathway", "q-value", "fold_enrichment", "cohort"]]
df1.to_csv("dem_pathways_kegg.csv", index=False)
```
